1) Сравните метрики hit_rate@k, precision@k. Какую матрику использовать предпочтительно и почему. Привидите пример 2-3 задач (опишите, что является клиентом, что товаром), в которой более уместно использовать метрику hit_rate?

Метрика hit_rate@k интересна, когда интересует точечное решение, precision@k целесообразна, 
когда интересует вариативность представленных релевантных решений. В большинстве случаев предпочтительней precision@k,
поскольку предполагает выбор решений для клиента, но в некоторых случаях важнее сам факт наличия этого решения. Например, мы хотим оценить деятельность социальной службы, которая должна предоставлять социальное желье,
отвечающее определенным критериям, гражданам, которые в нем нуждаются или имеют на него право. 
Другой случай, когда речь идет, например, о спасении жизни больных,
нас в первую очередь интересует был ли предложен вариант лечения, позволяющий предполагать положительный 
результат с высокой степенью вероятности. 
Клиентами являются гражданин и больной, товарами - жилье и жизнь.


2)В метрике NDCG@k мы используем логарифм в знаменателе. Как Вы думаете, почему именно логарифм? Какую функцию можно использовать вместо логарифма? Привидите пример метрик/подходов к предобработке данных/функций ошибок в ML, где также в знаменателе присутствует логарифм.

Функция логарифма вначале (X>0) растет быстрее, чем на "переферии", что позволяет отмасштабировать ряд таким образом,
чтобы установить большие расстояния между первыми элементами, чем последними, тем самым выделив топ. В качестве альтернативы
можно рассмотреть сигмоиду или гиперболический тангенс, которые используются в активаторах сетей.
